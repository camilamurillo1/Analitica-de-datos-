**FILTROS BASICOS CON CONETORES LOGICOS**

```python
#Filtro Y "&
filtro7=df[(df["PRECIO_UNITARIO"] < 6000) & (df["NOMBRE_VENDEDOR"] == "DIEGO ARMANDO LEON DURAN")]
filtro7
```

```python
filtro7.to_csv("Filtro7.csv")

from google.colab import files 
files.download("Filtro7.csv")

#se filtraron para que aparezcan los precios unitarios menores a 6.000 del cliente Diego Armando Leon
```

```python
#Filtro o "|"
filtro8=df[(df["COSTO_UNITARIO_CALCULADO"] < 500) | (df["NOMBRE_VENDEDOR"] == "ALFREDO CANELA")]
filtro8
```

```python
filtro8.to_csv("Filtro8.csv")

from google.colab import files 
files.download("Filtro8.csv")

#se filtraron para que aparezcan los costos unitarios calculados menores a 1000 o bien los datos del vendedor Alfredo Canela 
```

```python
#Filtro not "&"
filtro9=df[~(df["MARGEN_TOTAL_CALCULADO"] <= 6000) & ~(df["FECHA_DOC"] == "2022-01-05")]
filtro9
```

```python
filtro9.to_csv("Filtro9.csv")

from google.colab import files 
files.download("Filtro9.csv")

#se filtraron para que se eliminen todos aquellos margenes calculados menores o igual a 6.000 y tampoco los que tienen fecha correspodiente a 2022-01-05
```
